```python
def imdb_people_id(namesimdb):
    url = f"https://api.themoviedb.org/3/find/{namesimdb}?api_key={api_key}&language=en-US&external_source=imdb_id"
    re = requests.get(url).json()['movie_results'][0]
    
    return {"name": re["original_title"]}
```

```python
except_ids = []
df_tmdb = pd.DataFrame(columns = ["name"])

for namesimdb in namesimdb:
    try:
        df_tmdb = df_tmdb.append(imdb_people_id(namesimdb),ignore_index=True)
    except:
        except_ids.append(namesimdb)
        
df_tmdb
```

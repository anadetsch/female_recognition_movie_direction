```python
import pandas as pd
# import pymysql
# import getpass
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
gender_awards = pd.read_pickle("directors_genders_awards.pkl")
```

```python
gender_awards.head(100)
```

```python
gender_awards.plot()
```

```python
gender_awards_year = (gender_awards
                      .groupby(["Year"])
                      .agg({"Winner": np.count_nonzero}))
```

```python
gender_awards_year.plot()
```

```python
gender_awards_new = gender_awards.groupby(["Year", "director_gender"]).agg({"Winner" : "sum"}).reset_index()
```

```python
fig, ax = plt.subplots(figsize=(16,9))

sns.lineplot(data = gender_awards_new,
            x = "Year",
            y = "Winner",
            hue = "director_gender");
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = "https://en.wikipedia.org/wiki/List_of_female_film_and_television_directors"
```

```python
wiki = requests.get(url).content
soup = BeautifulSoup(wiki, "html.parser")
```

```python
results = (soup
           .find_all("div", class_ = "div-col columns column-width"))
```

```python
name_clean = [name
              .text
              .split("\n") for name in results]
```

```python
names_list = [name for sublist in name_clean for name in sublist]
names_list

for name in names_list:
    if name == '':
        names_list.remove(name)
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
import numpy as np
import json
import matplotlib
from pandas import json_normalize

api_key = "18cbd7f73d354ecf156bc195fa130139"
```

```python
url = "https://en.wikipedia.org/wiki/List_of_film_and_television_directors"
```

```python
wiki = requests.get(url).content
soup = BeautifulSoup(wiki, "html.parser")
soup
```

```python
results = (soup
           .find_all("div", class_ = "div-col columns column-width"))


results
```

```python
name_clean = [name
              .text
              .split("\n") for name in results]
name_clean
```

```python
type(name_clean)
```

```python
names_list = [name for sublist in name_clean for name in sublist]
names_list
```

```python
for name in names_list:
    if name == '':
        names_list.remove(name)
```

```python
names_list
```

```python
len(names_list)
```

```python
# Bringing names into format (whitspaces)

name = [x.replace(" ","%20") for x in names_list]
```

```python
# Creating empty Dataframes

df_all = pd.DataFrame()
df_clean = pd.DataFrame()
error_list = []
```

```python
for x in name:
    try:
        url = f"https://api.themoviedb.org/3/search/person?api_key={api_key}&language=en-US&query={x}&page=1&include_adult=false"
        rdata = requests.get(url).json()
        df_all = df_all.append(pd.DataFrame(json_normalize(rdata["results"], ["known_for"], ["name", "gender", "popularity", "id"] , meta_prefix="director_", record_prefix="movie_")), ignore_index=True)
    except:
        error_list.append(x)
        
df_all
```
